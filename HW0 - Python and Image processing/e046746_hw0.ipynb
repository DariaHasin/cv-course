#### Loading the Data
A dataset is a dictionary-like object that holds all the data and some metadata about the data. This data is stored in the `.data` member, which is a `n_samples, n_features` array. In the case of supervised problem, one or more response variables are stored in the `.target` member. More details on the different datasets can be found in the dedicated section.

```python
# import dataset
from sklearn import datasets
digits = datasets.load_digits()
```

For instance, in the case of the digits dataset, `digits.data` gives access to the features that can be used to classify the digits samples:

```python
print(digits.data)
```

and `digits.target` gives the ground truth for the digit dataset, that is the number corresponding to each digit image that we are trying to learn:

```python
print(digits.target)
```

Shape of the data arrays

The data is always a 2D array, shape `(n_samples, n_features)`, although the original data may have had a different shape. In the case of the digits, each original sample is an image of shape `(8, 8)` and can be accessed using: `digits.images`

```python
fig = plt.figure(figsize=(8, 5)) # create a figure, just like in matlab
ax = fig.add_subplot(1, 1 ,1) # create a subplot of certain size
ax.imshow(digits.images[0], cmap='gray')
ax.set_title("digits.image[0]")
```

#### Learning and Predicting

In the case of the digits dataset, the task is to predict, given an image, which digit it represents. We are given samples of each of the 10 possible classes (the digits zero through nine) on which we fit an estimator to be able to predict the classes to which unseen samples belong.

In scikit-learn, an estimator for classification is a Python object that implements the methods `fit(X, y)` and `predict(T)`.

```python
from sklearn.model_selection import train_test_split
# split to train and test sets
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
# fit a Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
# train
clf.fit(X_train, y_train)
# predict
y_pred = clf.predict(X_test)
# cacluclate accuracy
print("prediction accuracy: {:.3f}%".format(np.mean(y_pred == y_test) * 100))
```

### <img src="https://img.icons8.com/color/96/000000/deadlift.png" style="height:50px;display:inline"> Exercise 8

We are now going to cover the theoretical background on image filters.  

* Create the following kernels and apply them on `oy.jpg` (via convolution)
  * Average filter (size: $5\times 5$)
  * Gaussian filter (size: $5\times 5$)
  * Prewitt filter (size: $3\times 3$)
  * Sobel filter (size: $3\times 3$)
  * Laplacian filter (size: $3\times 3$)
  * Use `signal.convolve2d` from `scipy` for the convolution
* Show the results and explain what each filter is doing.
  * Show and explain how the $\sigma$ parameter of the gaussian filter affects the results.
* What is the difference between full/valid/same convolutions? Show the difference in results on one of the previous filters.
* What is the difference between Prewitt and Sobel filters?

```python
def gaussian_kernel(size,sigma):
    '''
    This function takes in a kernel size (int type) and sigma (float type)
    and returns numpy array of a Gaussian kernel for the wanted size and sigma.
    '''
    mu = np.floor([size // 2, size // 2])
    gauss_kernel = np.zeros((size, size))
    if(sigma != 0):
        for i in range(size):
            for j in range(size):
                gauss_kernel[i, j] = np.exp(-(0.5/(sigma**2)) * (np.square(i-mu[0]) + np.square(j-mu[0]))) / np.sqrt(2*np.pi*sigma**2)
    else:
        gauss_kernel = np.array([1, 4, 6, 4, 1], dtype=np.float64).reshape(5, 1)
        gauss_kernel = gauss_kernel @ gauss_kernel.T
    
    gauss_kernel /= gauss_kernel.sum()
    
    return gauss_kernel
```

```python
from scipy.signal import convolve2d

average_filter = np.ones((5,5), np.float32)/25
gaussian_filter = gaussian_kernel(5,0) #sigma=0
prewitt_filter = [[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]
sobel_filter = [[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]
laplacian_filter = [[0, -1, 0], [-1, 4, -1], [0, -1, 0]]
```

```python
def filter_img(rgb_img, filter_mat, mode='full'):
    filtered_img = np.stack((convolve2d(rgb_img[:, :, 0], filter_mat, mode=mode), 
                             convolve2d(rgb_img[:, :, 1], filter_mat, mode=mode), 
                             convolve2d(rgb_img[:, :, 2], filter_mat, mode=mode)), axis=2)
    return filtered_img.astype(np.uint8)
```

```python
fig = plt.figure(figsize=(25, 25))
filter1 = fig.add_subplot(5, 1 ,1)
filter1.imshow(filter_img(img_oy_rgb, average_filter))
filter1.axis('off')
filter1.set_title('Average filter')

filter2 = fig.add_subplot(5, 1 ,2)
filter2.imshow(filter_img(img_oy_rgb, gaussian_filter))
filter2.axis('off')
filter2.set_title('Gaussian filter')

filter3 = fig.add_subplot(5, 1 ,3)
filter3.imshow(filter_img(img_oy_rgb, prewitt_filter))
filter3.axis('off')
filter3.set_title('Prewitt filter')

filter4 = fig.add_subplot(5, 1 ,4)
filter4.imshow(filter_img(img_oy_rgb, sobel_filter))
filter4.axis('off')
filter4.set_title('Sobel filter')

filter5 = fig.add_subplot(5, 1 ,5)
filter5.imshow(filter_img(img_oy_rgb, laplacian_filter))
filter5.axis('off')
filter5.set_title('Laplacian filter')

plt.show()
```

### Show the results and explain what each filter is doing:
**Average filter** - Smoothing the image by reducing the amount of intensity variation between neighbouring pixels.
***
**Gaussian filter** - Reducing image noises and details by smooting the image.
***
**Prewitt filter** - Detects edges in the image. It can detect two types of edges, depend on the selected kernel - in this case horizontal edges.
***
**Sobel filter** - Detects edges in the image, by using image first order derivative in one axis and image average in the other axis which has the advantage of providing derivatives (which gives the edge detection) and smoothing (which reduces noise) concurrently. It can detect two types of edges, depend on the selected kernel - in this case horizontal edges.
***

* For $X_i \sim \mathcal{N}(\mu_0,\,\sigma^{2})$ and given IID observations- $X_1,...,X_n$, find the maximum likelihood estimator (MLE) of $\sigma$.
    
---
* What is the main difference between MLE and MAP estimators?

### Maximum likelihood estimator of $\sigma$
let $\hat{\sigma}_{MLE} = argmax_\sigma(L) \approx argmax_\sigma ln(L)$ while $L=\Pi_{i=1}^n f(x_i|\sigma)$ is the likelihood function.<br/><br/>
Therefore $\hat{\sigma}_{MLE} = argmax_\sigma ln(\Pi_{i=1}^n f(x_i|\sigma))$ <br/><br/>
To get the argmax we need to differentiate and compare to zero: 
$\frac{\partial}{\partial\sigma}ln(\Pi_{i=1}^n f(x_i|\sigma))=0$ <br/><br/>
$\frac{\partial}{\partial\sigma}ln(\Pi_{i=1}^n f(x_i|\sigma))=\frac{\partial}{\partial\sigma}\sum_{i=1}^n ln(f(x_i|\sigma))=\frac{\partial}{\partial\sigma}\sum_{i=1}^n ln(\frac{1}{\sigma \sqrt{2\pi}} \exp^{\frac{-(x_i-\mu_0)^2}{2\sigma^2}})=\frac{\partial}{\partial\sigma}\sum_{i=1}^n[-\frac{1}{2}ln(2\pi\sigma^2)-\frac{(x_i-\mu_0)^2}{2\sigma^2}]=0$ <br/><br/> $\Rightarrow -\frac{n}{\sigma}+\frac{1}{\sigma^3}\sum_{i=1}^n(x_i-\mu_0)^2=0$ <br/><br/>
$\Rightarrow n\sigma^2=\sum_{i=1}^n(x_i-\mu_0)^2$<br/><br/>
Therefore $\hat{\sigma}_{MLE}=\sqrt{\frac{1}{n}\sum_{i=1}^n(x_i-\mu_0)^2}$
  
___
### What is the main difference between MLE and MAP estimators?

Both methods give a single estimator value. MLE gives a value which maximises the likelihood $P(D|\theta)$, while MAP includes prior in the formula so it gives a value which maximises the posterior probability $P(\theta|D)$.
